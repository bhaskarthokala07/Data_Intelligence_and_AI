# Student Performance Management System

```python
students = []

# Function to calculate grade
def calculate_grade(avg):
    if avg >= 75:
        return "Distinction"
    elif avg >= 60:
        return "First Class"
    elif avg >= 50:
        return "Second Class"
    else:
        return "Fail"

# Add Student
def add_student():
    roll = int(input("Enter Roll Number: "))
    name = input("Enter Name: ")
    marks = []

    for i in range(3):
        mark = float(input(f"Enter Marks of Subject {i+1}: "))
        marks.append(mark)

    avg = sum(marks) / len(marks)
    grade = calculate_grade(avg)

    students.append({
        "roll": roll,
        "name": name,
        "marks": marks,
        "average": avg,
        "grade": grade
    })

    print("Student added successfully!")

# View Students
def view_students():
    if not students:
        print("No records found.")
        return

    for s in students:
        print("\nRoll No:", s["roll"])
        print("Name:", s["name"])
        print("Marks:", s["marks"])
        print("Average:", round(s["average"], 2))
        print("Grade:", s["grade"])

# Search Student
def search_student():
    choice = input("Search by Roll Number (R) or Name (N): ").upper()

    if choice == "R":
        roll = int(input("Enter Roll Number: "))
        for s in students:
            if s["roll"] == roll:
                print(s)
                return

    elif choice == "N":
        name = input("Enter Name: ").lower()
        for s in students:
            if s["name"].lower() == name:
                print(s)
                return

    print("Student not found!")

# Update Student
def update_student():
    roll = int(input("Enter Roll Number to Update: "))

    for s in students:
        if s["roll"] == roll:
            s["name"] = input("Enter New Name: ")

            marks = []
            for i in range(3):
                mark = float(input(f"Enter New Marks for Subject {i+1}: "))
                marks.append(mark)

            s["marks"] = marks
            s["average"] = sum(marks) / len(marks)
            s["grade"] = calculate_grade(s["average"])

            print("Record updated successfully!")
            return

    print("Student not found!")

# Delete Student
def delete_student():
    roll = int(input("Enter Roll Number to Delete: "))

    for s in students:
        if s["roll"] == roll:
            students.remove(s)
            print("Record deleted successfully!")
            return

    print("Student not found!")

# Class Average
def class_average():
    if not students:
        print("No student records available.")
        return

    total = sum(s["average"] for s in students)
    print("Class Average =", round(total / len(students), 2))

# Assign Grades
def assign_grades():
    for s in students:
        s["grade"] = calculate_grade(s["average"])

    print("Grades assigned successfully!")

# Filter by Grade
def filter_grade():
    grade = input("Enter Grade Category: ")

    found = False
    for s in students:
        if s["grade"].lower() == grade.lower():
            print(s)
            found = True

    if not found:
        print("No students found in this category.")

# Find Topper
def find_topper():
    if not students:
        print("No student records available.")
        return

    highest = max(s["average"] for s in students)

    print("\nClass Topper(s):")
    for s in students:
        if s["average"] == highest:
            print("Roll No:", s["roll"],
                  "Name:", s["name"],
                  "Average:", round(s["average"], 2))

# Main Menu
while True:
    print("\n===== Student Performance Management System =====")
    print("1. Add New Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student Record")
    print("5. Delete Student")
    print("6. Calculate Class Average")
    print("7. Assign Grades to All Students")
    print("8. Filter Students by Grade")
    print("9. Find Class Topper")
    print("10. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_student()
    elif choice == 2:
        view_students()
    elif choice == 3:
        search_student()
    elif choice == 4:
        update_student()
    elif choice == 5:
        delete_student()
    elif choice == 6:
        class_average()
    elif choice == 7:
        assign_grades()
    elif choice == 8:
        filter_grade()
    elif choice == 9:
        find_topper()
    elif choice == 10:
        print("Exiting Program...")
        break
    else:
        print("Invalid Choice!")
```
